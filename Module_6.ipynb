```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader.data as pdr
from datetime import datetime,timedelta
```

```python
start = datetime(2019,8,25)
end = datetime(2020,8,25)

df = pdr.DataReader(['ASIANPAINT.NS','AXISBANK.NS','BAJFINANCE.NS','BPCL.NS','CIPLA.NS','DRREDDY.NS','HDFCBANK.NS','INFY.NS','ITC.NS','SUNPHARMA.NS','ASHOKA.NS','FORTIS.NS','IDFC.NS','NCC.NS','PVR.NS','RAYMOND.NS','SUZLON.NS','VIPIND.NS','BOMDYEING.NS','SONATSOFTW.NS','BERGEPAINT.NS','DHFL.NS','GODREJIND.NS','IDBI.NS','IGL.NS','MINDTREE.NS','MRPL.NS','RBLBANK.NS','TATAPOWER.NS','VOLTAS.NS'],'yahoo',start,end)
df2 = df['Close']
df2.head()
```

```python
df2.columns
```

```python
returns = df2.pct_change().mean() * 252
returns = pd.DataFrame(returns)
returns.columns = ['Returns']
returns['Volatility'] = df2.pct_change().std() * np.sqrt(252)
returns
```

```python
from sklearn.cluster import KMeans

list = []
for k in range(1, 20):
    k_means = KMeans(n_clusters=k)
    k_means.fit(returns)
    list.append(k_means.inertia_)
```

```python
frame = pd.DataFrame({ 'Cluster' :range(1,20) , 'list':list })
```

```python
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'],frame['list'],marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
returns_scaled = scaler.fit_transform(returns)
```

```python
pd.DataFrame(returns_scaled).describe()
```

```python
list_scaled = []
```

```python
for k in range(1, 20):
    k_means = KMeans(n_clusters=k)
    k_means.fit(returns_scaled)
    list_scaled.append(k_means.inertia_)
```

```python
plt.figure(figsize=(12,6))
frame_scaled = pd.DataFrame({ 'Cluster' :range(1,20) , 'list':list_scaled })
plt.plot(frame_scaled['Cluster'],frame_scaled['list'],marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow curve')
```

```python
# optimum number of clusters are 4

k_means = KMeans(n_clusters = 4)
k_means.fit(returns_scaled)
pred = k_means.predict(returns_scaled)
pred
```

```python
frame = pd.DataFrame(returns_scaled)
frame['cluster'] = pred
frame['cluster'].value_counts()
```

```python
# To create clusters we need 4 steps

# Step-1:Choose the number of clusters and crate random centroid for each cluster
# Step 2 - Assign all the points to the closest cluster centroid
# Step 3 - Recompute centroids of newly formed clusters
# Step 4 - Repeat step 2 and 3.
```

```python
Centroids = (returns.sample(n=4))
fig, ax = plt.subplots()
ax.scatter(ret, vol)
ax.scatter(Centroids['Returns'],Centroids['Volatility'],c='red')
plt.gcf().set_size_inches((15, 10))

for i, txt in enumerate(col):
    ax.annotate(txt, (ret[i], vol[i]))
```

```python
diff = 1
j = 0
```

```python
# recomputing the centroids untill the value doesn't change

while(diff!=0):
    XD=returns
    i=1
    for index1,row_c in Centroids.iterrows():
        ED=[]
        for index2,row_d in XD.iterrows():
            d1=(row_c["Returns"]-row_d["Returns"])**2
            d2=(row_c["Volatility"]-row_d["Volatility"])**2
            d=np.sqrt(d1+d2)
            ED.append(d)
        returns[i]=ED
        i=i+1

    C=[]
    for index,row in returns.iterrows():
        min_dist=row[1]
        pos=1
        for i in range(4):
            if row[i+1] < min_dist:
                min_dist = row[i+1]
                pos=i+2
        C.append(pos)
    returns["Cluster"]=C
    Centroids_new = returns.groupby(["Cluster"]).mean()[["Volatility","Returns"]]
    if j == 0:
        diff=1
        j=j+1
    else:
        diff = (Centroids_new['Volatility'] - Centroids['Volatility']).sum() + (Centroids_new['Returns'] - Centroids['Returns']).sum()
        print(diff.sum())
    Centroids = returns.groupby(["Cluster"]).mean()[["Volatility","Returns"]]
```
