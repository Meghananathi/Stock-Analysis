```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("module1.csv")
```

```python
# Checking the type of date column

df.dtypes
```

```python
# Changing the date column datatype from 'object' to 'datetime64(ns)'

df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
df.head(3)
```

```python
# changing date column as index of the dataframe

df = df.set_index(df.Date)
df.index
```

```python
df.head(3)
```

```python
# dropping date column

df.drop(columns='Date',inplace=True)
df.head(3)
```

```python
# plot
plt.figure(figsize=(16,6))
plt.plot(df['ClosePrice'])
plt.xlabel('Date')
plt.ylabel('Close Price')
```

```python
# Stem-plot for daily change of stock price in percentage

plt.figure(figsize=(8,4))
plt.stem(df.index,df.Day_perc_change)
```

```python
# Renaming

df.rename(columns={'Total Traded Quantity':'Total_Traded_Quantity'}, inplace=True)
```

```python
df.columns
```

```python
# plotting volumes

sns.set(style="darkgrid")
plt.figure(figsize=(12,6))
plt.plot(df['Total_Traded_Quantity'])
plt.xlabel('Date')
plt.ylabel('Total_Traded_Quantity')
```

```python
# Stem-plot

plt.figure(figsize=(16,8))
plt.stem(df.index,df.Day_perc_change)
```

```python
# Relation between plotting and daily percentage change

sns.set(style="darkgrid")
scaledvolume =  df["Total_Traded_Quantity"] - df["Total_Traded_Quantity"].min()
scaledvolume = scaledvolume/scaledvolume.max() * df.Day_perc_change.max()

fig, ax = plt.subplots(figsize=(12, 6))

ax.stem(df.index, df.Day_perc_change , 'b', markerfmt='bo', label='Percente Change')
ax.plot(df.index, scaledvolume, 'k', label='Volume')
ax.set_xlabel('Date')
plt.legend(loc=0)
plt.tight_layout()
plt.xticks(plt.xticks()[0], df.index.date, rotation=45)
```

```python
from datetime import datetime ,timedelta
import pandas_datareader.data as pdr
```

```python
start = datetime(2017 , 5 ,16)
end = datetime(2019, 5 , 13)

## Reading data and saving only Close Price in other data frame.
data = pdr.DataReader(['ALLCARGO.NS','SUZLON.NS','MARUTI.NS','BAJFINANCE.NS','JUBILANT.NS','^NSEI'],'yahoo',start,end)
db = data.Close
db.head(3)
```

```python
df_percentage_change = db.pct_change()
```

```python
df_percentage_change.dropna(inplace=True)
df_percentage_change.head(3)
```

```python
## Now we have Close Price for all of the Above we Plot with Pair plot

sns.pairplot(df_percentage_change)
```

```python
db['^NSEI'] = db['^NSEI'].fillna((db['^NSEI'].mean()))
```

```python
# Plotting the mean for Niffty(^NSEI)

db['^NSEI'].rolling(window=7).mean().plot()
```

```python
# Plotting the standard deviation

db['^NSEI'].rolling(window=7).std().plot()
```

```python
# Calculating Volatality

volatility = db['^NSEI'].rolling(7).std() * np.sqrt(7)
volatility.dropna(inplace=True)
volatility.head(3)
```

```python
# Plotting Volatality

volatility.plot()
```

```python
# Moving averages for ^NSEI

start_date = '2017-05-16'
end_date = '2019-05-13'

short_rolling = db1.rolling(window=21).mean()
long_rolling = db1.rolling(window=34).mean()

fig, ax = plt.subplots(figsize=(16,9))

ax.plot(db1.loc[start_date:end_date, :].index, db1.loc[start_date:end_date, '^NSEI'])
ax.plot(long_rolling.loc[start_date:end_date, :].index, long_rolling.loc[start_date:end_date, '^NSEI'], label = '31_SMA')
ax.plot(short_rolling.loc[start_date:end_date, :].index, short_rolling.loc[start_date:end_date, '^NSEI'], label = '24_SMA')

ax.legend(loc='best')
ax.set_ylabel('Price in rupees')
```

```python
short_rolling = db1.rolling(window=21).mean()
long_rolling = db1.rolling(window=34).mean()

posn = long_rolling - short_rolling
mk = pd.DataFrame(index=posn.index)

mask = ((posn<0) & (posn>0)).all(axis=1)
mk['call'] = np.where(mask , 'Buy', 'Sell')

print('Buy on dates',mk[mk['call']=='Buy'])
print('Sell on dates',mk[mk['call']=='Sell'])
```

```python
# Moving averages of closing prices for 24day and 34day

short_rolling_df = df.ClosePrice.rolling(window=21).mean()
long_rolling_df = df.ClosePrice.rolling(window=34).mean()

posn_df = long_rolling_df - short_rolling_df
mk_df = pd.DataFrame(index=posn_df.index)

mask_df = ((posn_df<0) & (posn_df>0)).all(axis=0)
mk_df['call'] = np.where(mask_df , 'Buy', 'Sell')

print('Buy on dates',mk_df[mk_df['call']=='Buy'])
print('Sell on dates',mk_df[mk_df['call']=='Sell'])
```

```python
# Bollinger Brands

average =  df.ClosePrice.rolling(14).mean()
stdd = df.ClosePrice.rolling(14).std()

upperband = average + 2*stdd
lowerband = average - 2*stdd

upperband.plot()
lowerband.plot()
average.plot()
df.ClosePrice.plot(Color='black')
```

```python
# Bollinger Brands for 3 DATASETS(^NSEI,SUZLON.NS,BAJFINANCE.NS)

average_db1 =  db1.rolling(14).mean()
stdd_db1 = db1.rolling(14).std()

upperband_db1 = average + 2*stdd
lowerband_db1 = average - 2*stdd

upperband_db1.plot()
lowerband_db1.plot()
average_db1.plot()
```

```python
df['Upperband'] = upperband.values
df['Lowerband'] = lowerband.values
df['Middle'] = average.values
df.to_csv('module2.csv')
```
