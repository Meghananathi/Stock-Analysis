```python
# Average price over the last N days here N=10

def average(N):
    df_days = df.tail(N)
    avg = df_days['Close Price'].mean()
    print("average : {}%".format(avg))
    return
average(10)
```

```python
# Calculation of Profit/Loss over last N days

def profit_loss_pct(N):
    df_n_days = df.tail(N)
    df['Profit_Loss'] = ((df_n_days["Open Price"] - df_n_days["Close Price"])/df_n_days["Close Price"])
    total_days = len(df['Profit_Loss'])
    calc_pnl = df['Profit_Loss'][total_days-N:].sum()
    if df["Profit_Loss"][N] < 0:
          print("Loss pct is: {:5.2f}%". format(df["Profit_Loss"][N]*100));
    else:
         print("Profit pct is : {:5.2f}%". format(df["Profit_Loss"][N]*100));
    return
profit_loss_pct(494)
```

```python
# adding a column 'Day_Perc_Change'  

df['Day_perc_change'] = df.ClosePrice.pct_change()
df.head()
```

```python
## As we Can Observe that first row has NaN value
## So we will Drop the row or we can assign it any number or 0  so that it does not effect result

df.dropna(inplace=True)
df.head(3)

## or
# df.loc[0,['Day_Perc_Change']] = 0
```

```python
# Here we are using np.select

conditions = [(df.Day_perc_change>-0.5) & (df.Day_perc_change<0.5) , 
            (df.Day_perc_change < 1) & (df.Day_perc_change >0.5),
            (df.Day_perc_change>1) & (df.Day_perc_change<3),
            (df.Day_perc_change<-0.5) & (df.Day_perc_change>-1),
            (df.Day_perc_change<-1) & (df.Day_perc_change>-3),
            (df.Day_perc_change>3) & (df.Day_perc_change<7),
            (df.Day_perc_change<-3) & (df.Day_perc_change>-7),
            (df.Day_perc_change>7),
            (df.Day_perc_change<-7)
]


choices =[
            "Slight or No Change",
            'Slight Positive',
            'Positive',
            'Slight Negative',
            'Negative',
            'Among Top Gainers',
            'Among Top Losers',
            'Bull run',
            'Bear Drop'
    
]

df['Trend'] = np.select(conditions,choices)
df.head(5)
```

```python
print('The Average Value is',df.groupby(df.Trend)['Total Traded Quantity'].mean(),'\n')
print('The Median Value is' , df.groupby(df.Trend)['Total Traded Quantity'].median())
```
